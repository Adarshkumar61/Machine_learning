```python
# you can increase your markersize, use: 'markersize' or m'ms:

x_pinn = np.array([2,3,5,7,8])
y_pinn = np.array([12,5,32,43, 38])
plt.plot(x_pinn, y_pinn,'*-r', ms = 20,)
plt.show()
```

```python
# you can also color the marker:
# for edge: 'markeredgecolor' or 'mec':
# for inner : 'markerfacecolor' or mfc:
#you can use both to color entire marker.
x_pin = np.array([2,3,5,7,8])
y_pin = np.array([12,5,32,43, 38])
plt.plot(x_pin, y_pin,'*-r', ms = 20, mec= 'r', mfc='k')
plt.show()
```

```python
# you can also change style of line separately by using : 'linestyle' or 'ls':
# and  you change color seperately :
x_points = np.array([2,4,6,8,10])
y_points = np.array([13,5,7,7, 32])
# plt.plot(x_points, y_points, ls=':', color = 'red', marker='*', ms=15, mec='k', mfc='r')
# doing together shortcut:
plt.plot(x_points, y_points, '*:r', ms=14, mec='b')
plt.show()
```

```python
# you can also change line width by: 'linewidth' or 'lw':
x_points = np.array([2,4,6,8,10])
y_points = np.array([13,5,7,7, 32])
plt.plot(x_points, y_points, lw=12)
plt.show()
```

```python
# use loc to change position of title, xlabel (default = center): 'loc' but for ylabel use : bottom, center or top:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
font1 = {'family': 'serif', 'color': 'red', 'size': 20}
font2 = {'family': 'serif', 'color': 'blue',  'size': 14}

plt.title('data', fontdict = font1, loc= 'left')
plt.xlabel('heart beat', fontdict = font2, loc = 'right')
plt.ylabel('calories burned', fontdict = font2, loc = 'bottom')
plt.show()
```

```python
# The scatter() function plots one dot for each observation.
# It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:
```

```python
# Example
# Draw two plots on the same figure:

x_point = np.array([0.1, 0.8, 12, 34, 43])
y_point = np.array([27, 76, 34, 23, 65])
plt.scatter(x_point, y_point, color= 'm')


xx_po = np.array([0.8, 3, 7, 9, 12])
yy_po = np.array([43, 21, 65, 15, 19])
plt.scatter(xx_po, yy_po)

third = np.array([0.4,32, 21, 67, 23])
fourth = np.array([43, 23, 45, 25, 12])
plt.scatter(third, fourth)
plt.show()
```

```python
# You can even set a specific color for each dot by using an array of colors as value for the c argument:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
# You can change the size of the dots with the s argument.

# Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) 
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

```python
# Alpha: you can control the transparancy of marker using 'alpha':

x_pi = np.array([2,3,5,6,6,7,87,8,45])
y_pi = np.array([3,2,45,6,5,7,76,43, 21])
plt.scatter(x_pi, y_pi, alpha = 0.3, c= 'red')
plt.show()
```

```python
# Creating our first model with scikit learn  model is linearregression:
```

```python
# Fitting data:
```

```python
x = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2,2], [1, 2.5]]) #input
y = np.array([0,0,0,1,1,1])  #output
```

```python
# Fitting the model:
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x, y)
```

```python
# you can make predition by calling .predict fn
```

```python
# now its time to make prediction:
y_predict = model.predict(x)
print('prediction on training set: ',y_predict)
```

```python
# you can check accuracy by calling ''.score' fn:

print('accuracy is: ', model.score(x, y))
```
